```python
# Integer Data Type
a = 12000
print(a)
print("The data type of a is : ",type(a))
```

```python
# Float Data Type
float1 = 3.145
print(float1)
print("The data type of float1 is : ", type(float1))
```

```python
# List Data Type
student_list = ["Pratik","Farhan","Adeeba"]
print(student_list)
print("The data type of list : ", type(student_list))
```

```python
# Tuple Data Type
rainbowColor = ("Red","Green","Blue","Orange","Indigo","Violet","Yellow")
print(rainbowColor)
print("The data type of rainbowColor is : ", type(rainbowColor))
```

```python
# Dictionary Data Type
student_info = {
    "name":"Pratik Honmane",
    "mobile_number":7517330553,
    "address":"Ahilyanagar"
}

print(student_info)
print("The data type of student_info is : ", type(student_info))
```

```python
# Set Data Type
A = {2,4,6,8,10}
print(A)
print("The data type of even_number is : ", type(A))
```
